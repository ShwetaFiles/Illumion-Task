```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import RectangleSelector

# Memory-map the data file for efficient access

movie_data = np.load('C:/Users/pilla/Downloads/movie.npy', mmap_mode='r')
time_data = np.load('C:/Users/pilla/Downloads/time.npy', mmap_mode='r')
print(f"Movie Data Shape: {movie_data.shape}")
print(f"Time Data Shape: {time_data.shape}")
```

```python
# Global variable for rectangle coordinates
rect_coords = None

# Function to handle rectangle selection and trigger intensity calculation
def on_select(eclick, erelease):
    global rect_coords
    rect_coords = (
        int(eclick.xdata), int(eclick.ydata),
        int(erelease.xdata), int(erelease.ydata)
    )
    print(f"Rectangle selected with coordinates: {rect_coords}")
#     plot_relative_intensity()  # Call this function after selection

# Initialize plot and show the first frame
fig, ax = plt.subplots()
img_display = ax.imshow(movie_data[0], cmap='gray')
plt.colorbar(img_display, ax=ax)
ax.set_title("Select a rectangle to calculate mean intensity")

# Rectangle selector for interactive selection
rect_selector = RectangleSelector(
    ax, onselect=on_select, drawtype='box', useblit=True,
    button=[1], interactive=True
)
print("RectangleSelector initialized.")

# Draw_idle ensures the canvas updates when interacting with the selector
fig.canvas.draw_idle()

plt.show()
print("Interactive plot shown.")
```
